```python
 Check for Skewness- Categorical-ANOVA
```

```python
y = data['Number_of_Vehicles']
```

```python
from scipy.stats import f_oneway

categorical_vars = ['Road_Type', 'Road_Surface_Conditions', 'Weather_Conditions', 'Light_Conditions']

for var in categorical_vars:
    groups = [y[data[var] == category] for category in data[var].unique()]
    f_stat, p_value = f_oneway(*groups)
    print(f"ANOVA for {var}: F-statistic = {f_stat:.2f}, p-value = {p_value:.4f}")
```

## Check for multicollinearity

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor 
```

# USING HEATMAP TO CHECK FOR Multicollinearity

```python
input_variables = [
    'Road_Type',
    'Speed_limit',
    'Light_Conditions',
    'Weather_Conditions',
    'Road_Surface_Conditions',
    'Urban_or_Rural_Area'
]
```

```python
target_variable = 'Number_of_Vehicles'
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = data[input_variables].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Input Variables')
plt.show()
```

# OBSERVATION:

Speed_limit and Urban_or_Rural_Area: There is a positive correlation of 0.63. This suggests that the speed limit tends to be higher in either urban or rural areas (depending on how the variable is coded).
Weather_Conditions and Road_Surface_Conditions: There is a moderate positive correlation of 0.48, implying that certain weather conditions tend to correlate with certain road surface conditions, which is logical.
Road_Type and Speed_limit: A negative correlation of -0.31 indicates that different types of roads (e.g., highways vs. city streets) tend to have varying speed limits.

# USING Variance Inflation Factor (VIF) TO FURTHER CHECK FOR MULTICOLINEARITY

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif_data = pd.DataFrame()
vif_data["Variable"] = input_variables
vif_data["VIF"] = [variance_inflation_factor(data[input_variables].values, i) for i in range(len(input_variables))]
```

```python
print("Variance Inflation Factor (VIF) for each input variable:")
print(vif_data)
```

# OBSERVATION:

â€¢	Speed_limit (VIF = 15.388836) and Urban_or_Rural_Area (VIF = 15.371432) both have very high VIF values, indicating significant multicollinearity. These variables are likely explaining much of the same variance in your model's dependent variable and could be distorting your regression results.
â€¢	Light_Conditions (VIF = 7.567508) also has a high VIF, which suggests it has considerable multicollinearity with other variables.
â€¢	Road_Type (VIF = 4.675463) is on the higher end but still below the usual threshold of 5 that might indicate serious concerns.
â€¢	Weather_Conditions (VIF = 2.904205) and Road_Surface_Conditions (VIF = 1.937674) have VIF values well within acceptable ranges, indicating minimal concerns with multicollinearity.
